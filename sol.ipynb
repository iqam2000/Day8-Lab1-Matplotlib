```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Q1: Create two charts using list and dataframe

```python
#Using List
saudi_airline_schedule = ['Sunday','Monday','Tuesday','Wednesday','Thursday', 'Friday']
Saudies = [10,20,20,40,10,25]
nonSaudies = [20,30,25,30,16,10]
  
X_axis = np.arange(len(saudi_airline_schedule))
  
plt.bar(X_axis - 0.10, Saudies, 0.20,    label = 'Saudi')
plt.bar(X_axis + 0.10, nonSaudies, 0.20, label = 'nonSaudi')
  
plt.xticks(X_axis, saudi_airline_schedule)
plt.xlabel("Flight day")
plt.ylabel("Passenger Number")
plt.title("Passenger Nationlities")
plt.legend()
plt.show()
```
